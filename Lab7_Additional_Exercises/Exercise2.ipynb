```python
import pandas as pd

df = pd.read_csv("spotify_top_1000_tracks[1].csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

student_name = "Jenny Guevarra"
student_id = "20254172"

id_num = int(student_id[-3:])

color1 = f"#{(id_num * 7) % 256:02x}{(id_num * 3) % 256:02x}{(id_num * 5) % 256:02x}"
color2 = f"#{(id_num * 2) % 256:02x}{(id_num * 9) % 256:02x}{(id_num * 4) % 256:02x}"
```

```python
import numpy as np

df['year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year

yearly = df.groupby('year').size()

cumulative = yearly.cumsum()

plt.figure(figsize=(10, 6))

plt.bar(yearly.index, yearly.values, color=color2, alpha=0.6, label="Yearly Releases")

plt.plot(cumulative.index, cumulative.values, color=color1, marker='o', label="Cumulative")

plt.title(f"{student_name} ({student_id})\nSong Release Trend")
plt.xlabel("Year")
plt.ylabel("Number of Songs")

plt.legend()
plt.grid(True)
plt.show()
```
